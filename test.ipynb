```python
def ContactPotential(r, epilon=12, sigma=0.8):
    V = epilon * ((sigma/r)**12 - (sigma/r)**10)

    return V

def ContactPotential_v2(r, epilon=12, sigma=0.8):
    V = epilon * (5 * (sigma/r)**12 - 6 * (sigma/r)**10)

    return V

def LJPotential(r, epilon=12, sigma=0.8):
    V = epilon * ((sigma/r)**12 - (sigma/r)**6)

    return V
```

```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
X = np.linspace(0.2,1.1,100)
X
```

```python
sigma = 0.6
Y_contact = [ContactPotential(x, sigma=sigma, epilon=45) for x in X]
plt.plot(X,Y_contact, label='Contact')

Y_contact_v2 = [ContactPotential_v2(x, sigma=sigma*1.1, epilon=3) for x in X]
plt.plot(X,Y_contact_v2, label='Contact_v2', color='black')

Y_LJ = [LJPotential(x, sigma=sigma) for x in X]
plt.plot(X,Y_LJ, label='LJ')
plt.ylim(-5,10)
plt.legend()
```
